```python
# ecncode categorical value

from sklearn.preprocessing import LabelEncoder

labelencode_Y = LabelEncoder()
raw_df.iloc[:,1] =labelencode_Y.fit_transform(raw_df.iloc[:,1].values)
```

```python
## Split data into independent (X) and depnedent (y) variable

X = raw_df.iloc[:,2:31]
y = raw_df.iloc[:,1]
```

```python
# spliting data into 70% and 30%

from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scaling data(Feature Scaling)

from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
X_train.shape
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#import logistic regression

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
#print model coefficient
print(model.coef_.tolist())
```

```python
print(model.intercept_)
```

```python
train_pred = model.predict(X_test)
```

```python
train_pred
```

```python
pd.value_counts(train_pred)
```

```python
model.classes_
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, train_pred)
```

```python
# using confusion matrix to check accuracy of mode
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, train_pred)

print(cm)
```

```python
from sklearn.metrics import precision_score, recall_score

precision_score(y_test, train_pred)
```

```python
recall_score(y_test, train_pred)
```
